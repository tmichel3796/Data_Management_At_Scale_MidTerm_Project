```python
import polars as pl
from toolz import pipe
import requests, zipfile, io
```

## CREATE THE LINKS TO ACCESS THE DATA
please run all and needed data will be downloaded into data folder. 

### RESPONDENT

```python
#creates links for resp


maybe_pad_month = lambda n: f"0{n}" if n < 10 else f"{n}"

(urls_resp :=
 pipe(range(2003, 2025),
      lambda L: [maybe_pad_month(n) 
                 for n in L
                ],
      lambda L:  [f"https://www.bls.gov/tus/datafiles/atusresp{dash}{year}.zip"
                   for year in L
                  for dash in ['_','-']
                 ],
 )

)
```

### ACTIVITY

```python
#create links for act

# this does  the addresses for all the act and resp link addressses, 
#yes there are extra addresses, but we'll use try catch blocks so that all the extra addresses are ignored
maybe_pad_month = lambda n: f"0{n}" if n < 10 else f"{n}"

(urls_act :=
 pipe(range(2003, 2025),
      lambda L: [maybe_pad_month(n) 
                 for n in L
                ],
      lambda L:  [f"https://www.bls.gov/tus/datafiles/atusact{dash}{year}.zip"
                   for year in L
                  for dash in ['_','-']
                 ],
 )

)
```

### WHO


```python
#make who links
#https://www.bls.gov/tus/datafiles/atuswho-2024.zip
#https://www.bls.gov/tus/datafiles/atuswho_2003.zip

maybe_pad_month = lambda n: f"0{n}" if n < 10 else f"{n}"

(urls_who :=
 pipe(range(2003, 2025),
      lambda L: [maybe_pad_month(n) 
                 for n in L
                ],
      lambda L:  [f"https://www.bls.gov/tus/datafiles/atuswho{dash}{year}.zip"
                   for year in L
                  for dash in ['_','-']
                 ],
 )

)
```

### ROSTER

```python
#find url's for the new roster files


maybe_pad_month = lambda n: f"0{n}" if n < 10 else f"{n}"

(urls_rost :=
 pipe(range(2003, 2025),
      lambda L: [maybe_pad_month(n) 
                 for n in L
                ],
      lambda L:  [f"https://www.bls.gov/tus/datafiles/atusrost{dash}{year}.zip"
                   for year in L
                  for dash in ['_','-']
                 ],
 )

)
```

### DATA DICTIONARY
Allows download of data dictionaries for every year.

```python
#creates url's for dictionaries

(urls_dict :=
 pipe(range(3, 25) ,
      lambda L: [maybe_pad_month(n) 
                 for n in L
                ] + [""],
      lambda L:  [f"https://www.bls.gov/tus/dictionaries/atuscpscodebook{year}.pdf"
                   for year in  L 
                 ],
 )

)
```

### INTERVIEW DICTIONARY

```python
# find url's for the interview dictionaries

#creates url's for dictionaries https://www.bls.gov/tus/dictionaries/atusintcodebk24.pdf
#https://www.bls.gov/tus/dictionaries/atusintcodebook.pdf

(urls_dict_int :=
 pipe(range(3, 25) ,
      lambda L: [maybe_pad_month(n) 
                 for n in L
                ] + [""],
      lambda L:  [f"https://www.bls.gov/tus/dictionaries/atusintcodebk{year}.pdf"
                   for year in  L 
                 ],
 )

)

```

## DOWNLOAD AND UNZIP FILES

### RESPONDENT 

```python
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"
}

for url in urls_resp:                                  # Make `some_urls` match above name
    try:
        print(f"Downloading {url}")
        r = requests.get(url, headers=headers)
        if r.ok:
            try:
                print(f"Unzipping {url}")
                z = zipfile.ZipFile(io.BytesIO(r.content))
                z.extractall("./data")
                print(f"Successfully downloaded and unzipped {url}\n\n")
            except zipfile.BadZipFile:
                print(f"Downloaded file is not a valid ZIP: {url}\n\n")
        else:
            print(f"Failed to download {url} with status code {r.status_code}\n\n")
    except Exception as e:
        print(f"Problem downloading {url}: {e}\n\n")
```

### DATA DICTIONARY

```python
for url in urls_dict:
    try:
        print(f"Downloading {url}")
        r = requests.get(url, headers=headers)
        if r.ok:
            try:
                filename = url.split("/")[-1]
                with open(f"./data/{filename}", "wb") as f:
                    f.write(r.content)
                print(f"Unzipping {url}")
                print(f"Successfully downloaded  {url}\n\n")
            except Exception as e:
                print(f"Downloaded file is not a valid PDF: {url}\n\n")
        else:
            print(f"Failed to download {url} with status code {r.status_code}\n\n")
    except Exception as e:
        print(f"Problem downloading {url}: {e}\n\n")
```
